```python
#13
def minimum_skew(genome):
    skew = 0
    skew_values = []
    
    for nucleotide in genome:
        if nucleotide == 'G':
            skew += 1
        elif nucleotide == 'C':
            skew -= 1
        skew_values.append(skew)
    
    min_skew = min(skew_values)
    min_positions = [i+1 for i, value in enumerate(skew_values) if value == min_skew]
    
    return min_positions 

# Example usage
with open('../rosalind_ba1f.txt', 'r') as file:
    string = file.read().strip()
result = minimum_skew(string)
print(result)
```
